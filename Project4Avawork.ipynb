```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np


Twilight_movies_clean = Twilight_movies.dropna(subset=['rating'])

Twilight_movies_clean['rating_class'] = pd.cut(Twilight_movies_clean['rating'], bins=[0, 1, 2, 3, 4, 5], labels=[1, 2, 3, 4, 5])


features = Twilight_movies_clean[['movieId', 'userId']]
target = Twilight_movies_clean['rating_class']


X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=10)


model = SVC(kernel='linear')
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print(predictions)
```
